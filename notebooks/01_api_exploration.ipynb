```python
import requests
import pandas as pd
```

```python
url_deputados = "https://dadosabertos.camara.leg.br/api/v2/deputados"

params_deputados = {
    "itens": 10,
    "pagina": 1,
    "ordem": "ASC",
    "ordenarPor": "nome"
}

response_deputados = requests.get(url_deputados, params=params_deputados)

print(response_deputados.status_code)

deputados_json = response_deputados.json()

deputados_json.keys()
```

```python
deputados_json["dados"][:2]
```

```python
df_deputados = pd.DataFrame(deputados_json["dados"])

df_deputados.head()
```

```python
df_deputados.info()
df_deputados.columns
```

```python
id_deputado = df_deputados.loc[0, "id"]

id_deputado
```

```python
url_despesas = f"https://dadosabertos.camara.leg.br/api/v2/deputados/{id_deputado}/despesas"

params_despesas = {
    "ano": 2026,
    "mes": 3,
    "itens": 100,
    "pagina": 1,
    "ordem": "ASC"
}

response_despesas = requests.get(url_despesas, params=params_despesas)

print(response_despesas.status_code)

despesas_json = response_despesas.json()

despesas_json.keys()
```

```python
len(despesas_json["dados"])
```

```python
df_despesas = pd.DataFrame(despesas_json["dados"])

df_despesas.head()
```

# Resumo da exploração da API

## Endpoint de deputados

- Status HTTP: 200
- Quantidade de registros retornados no teste: 10
- Campos principais: id, uri, nome, siglaPartido, uriPartido, siglaUf, idLegislatura, urlFoto, email
- Observações: endpoint validado com sucesso e será usado para obter os IDs dos deputados.

## Endpoint de despesas

- ID de deputado testado: preencher com o ID usado
- Ano/mês testado: 2026/03
- Status HTTP: 200
- Quantidade de registros retornados: preencher com o total retornado
- Campos principais: preencher com as colunas retornadas
- Observações: endpoint validado com sucesso para o período de março de 2026.

## Decisões para a pipeline

- A extração de deputados será usada como base para obter IDs.
- A extração de despesas será feita por deputado, ano e mês.
- O período inicial do MVP será março de 2026.
- A paginação será obrigatória.
- A camada Bronze salvará o JSON bruto.
- A camada Silver será derivada a partir dos arquivos Bronze.
